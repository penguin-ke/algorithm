判斷輸入的數字是否為3的倍數(範圍1-300)

擷取.PNG

**時間換空間**

```
n=int(input('請輸入整數,範圍1-300 : '))
if n<10:               #判斷數字(若為1-9)
  if n%3==0:            #如果除3餘數為0,是三的倍數
    print(n,',是3的倍數')
  else:
    print(n,',不是3的倍數')
elif n>=10 and n<100:       #判斷數字(若為10-99)
  a1=n%10             #取得個位數
  b1=n/10             
  c1=int(b1)            #取得十位數
  if (a1+c1)%3==0:        #數字相加,如果除3餘數為0,是三的倍數
    print(n,',是3的倍數')
  else:
    print(n,',不是3的倍數')
elif n>=100:            #判斷數字(若為100-300)
  a2=n%10             #取得個位數
  b2=n/10
  c2=int(b2)
  d2=c2%10            #取得十位數
  e2=c2/10
  f2=int(e2)           #取得百位數
  if (a2+d2+f2)%3==0:      #數字相加,如果除3餘數為0,是三的倍數
    print(n,',是3的倍數')
  else:
    print(n,',不是3的倍數')



    

```
