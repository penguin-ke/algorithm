2.插入排序法(由小到大)

```
import random

def emergedata(data): #資料輸出
  for i in range(8):
    print(' %d'%data[i],end='') #數字之間以空格相間,不會是換行呈現 ( end 是當資料輸出結束時所插入的字元)

def select(data): #選擇排序法主要程式處
  for i in range(7):
    x=i+1 #給定後面第幾次排序,所設的值
    min=i #給定要比較的數值其索引值 (比較後,以索引值改變的方式)
    for j in range(i+1,8):
      if data[min]<data[j]: #min位的數值依序與後面的數值做比較
        min=j  #上述若成立,min的索引值變為被比較數值的索引值
    data[i],data[min]=data[min],data[i]  #一輪比較結束後,min所對應的數值會是比較後的最大值,將min的值與剛開始比較的數值對調
    print("第%d次排序:"%x,end='')
    emergedata(data) #印出每一輪比較完後的結果
    print('\n')
        
data=list(range(1,101)) #建立一個串列(包含數字1-100)
random.shuffle(data)  #將串列裡的數值打亂
figure=data[:]     #複製data串列給下方插入排序法用 (串列彼此不會影響)

print("選 擇 排 序\n")
print('原始數列 :',end='')
emergedata(data) #印出剛開始的8個亂數
print('\n-----------------------------------')
select(data)
print()


def showdata(figure): #資料輸出
  for i in range(8):
    print(' %d'%figure[i],end='')  #數字之間以空格相間,不會是換行呈現 ( end 是當資料輸出結束時所插入的字元)

def insert(figure):  #插入排序法主要程式處
  for i in range(1,8):
    tmp=figure[i] #用來暫存資料
    no=i-1 
    while no>=0 and tmp<figure[no]:
      figure[no+1]=figure[no] #上述若成立,把所有元素往後推一個位置
      no-=1
    figure[no+1]=tmp #最小的元素放到第一個元素
    print("第%d次排序:"%i,end='')
    showdata(figure) #印出每一輪比較完後的結果
    print('\n')

print("插 入 排 序\n")
print('原始數列 :',end='')
showdata(figure) #印出剛開始的8個亂數
print('\n-----------------------------------')
insert(figure)
```
