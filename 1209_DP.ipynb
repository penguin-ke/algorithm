```
# 自底向上
def BottomUpCutRod(p, x):
    r = [0]*(x+1)
    for n in range(1, x+1):
        if x == 0:
            return 0
        q =0
        for i in range(1, n+1):
            q = max(q, p[i]+r[n-i])
            r[n] = q
            #print(r)
        #print(q)
    print("最大收益為:",r[n])
p=[0,1,5,8,9,10,17,17,20,24,30] #對應長度的價格表
x=int(input("請輸入鐵條長度:"))
BottomUpCutRod(p, x)   #  (30, [0, 1, 5, 8, 10, 13, 17, 18, 22, 25, 30])
```
