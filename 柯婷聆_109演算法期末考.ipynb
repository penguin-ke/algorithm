解題想法：

*程式碼為上網找尋,後改動一部分


1.dfs
起始節點往下從左而右找尋相鄰節點

若找尋的節點下面仍有節點,先拜訪下面

拜訪完後,回來再換另一個節點



2.bfs
起始節點往下從左而右找尋相鄰節點

若找尋的節點下面仍有節點,不先拜訪下面

是先拜訪同階層的節點


```
#程式
def enqueue(a, o):
    a.insert(0, o)

def dequeue(a):
    return a.pop()

g = { #  graph: 被搜尋的網路
    '1': {'n':['2','3'], 'v':0}, #  n: neighbor (鄰居), v: visited (是否被訪問過)
    '2': {'n':['4','5'], 'v':0},
    '3': {'n':['6'], 'v':0},
    '4': {'n':[], 'v':0},
    '5': {'n':['8'], 'v':0},
    '6': {'n':['9'], 'v':0},
    '7': {'n':[], 'v':0},
    '8': {'n':['7'], 'v':0},
    '9': {'n':[], 'v':0}
}

def init(g): #  初始化、設定 visited 為 0
    for i in g:
        g[i]['v'] = 0

def dfs(g, node): #  深度優先搜尋
    if g[node]['v']!=0:           #  如果已訪問過，就不再訪問
        return
    print(node, '=> ', end = '')     #  否則、印出節點
    g[node]['v'] = 1            #    並設定為已訪問
    neighbors = g[node]['n']        # 取出鄰居節點
    for n in neighbors:          #  對於每個鄰居
        dfs(g, n)            #    逐一進行訪問

queue=['1'] #  BFS 用的 queue, 起始點為 1。

def bfs(g, q): #  廣度優先搜尋
    if len(q)==0:                 #  如果 queue 已空，則返回。
        return
    node = dequeue(q)              #  否則、取出 queue 的第一個節點。
    if g[node]['v'] == 0:           #  如果該節點尚未拜訪過。
        g[node]['v'] = 1          #    標示為已拜訪
    else:                    #  否則 (已訪問過)
        return                #    不繼續搜尋，直接返回。
    print(node, '=> ', end = '')      #  印出節點
    neighbors = g[node]['n']          #  取出鄰居。
    for n in neighbors: #  對於每個鄰居
        if not g[n]['v']:         #  假如該鄰居還沒被拜訪過
            enqueue(q, n)        #    就放入 queue 中
    bfs(g, q)
    bfs(g, q)
print('dfs:', end = '')
init(g)
dfs(g, '1') # 呼叫深度優先搜尋。
print('')

print('bfs:', end = '')
init(g)
bfs(g, queue) # 呼叫廣度優先搜尋。
print('')
```

解題想法:

*程式碼為上網找尋,後改寫

1.以我自己理解的想法:一開始選擇"0"當作起始點,之後去尋找和"0"相連接的所有頂點

選擇最小的權重(邊),將其頂點加入(會印出其對應的邊)

2.後來的步驟也是再找尋已有的頂點其相鄰的所有頂點,選擇最小的權重(邊),將其頂點加入(會印出其對應的邊)

3.圖在程式中的寫法是以相鄰矩陣的方式

4.如果相鄰的頂點已有加入,則不考慮其邊

```
#程式
# Prim's Algorithm in Python


INF = 99999
V = 8 #圖中頂點數
# create a 2d array of size 5x5
# for adjacency matrix to represent graph
G = [[0, 30, 23, 0, 0, 16, 19, 0],
     [30, 0, 0, 0, 0, 0, 35, 32],      #相鄰矩陣
     [23, 0, 0, 20, 15, 26, 21, 0],
     [0, 0, 20, 0, 24, 0, 14, 64],
     [0, 0, 15, 24, 0, 25, 0, 2],
     [16, 0, 26, 0, 25, 0, 0, 0],
     [19, 35, 21, 14, 0, 0, 0, 40],
     [0, 32, 0, 64, 2, 0, 40, 0]]
# create a array to track selected vertex (創建一個數組以追蹤選定的頂點)
# selected will become true otherwise false (選擇的將為true，否則為false)
selected = [0, 0, 0, 0, 0, 0, 0, 0]
# set number of edge to 0
no_edge = 0
# the number of egde in minimum spanning tree will be
# always less than(V - 1), where V is number of vertices in #最小生成樹的邊數<節點數
# graph
# choose 0th vertex and make it true
selected[0] = True
# print for edge and weight #印出是哪條邊,及它的權重
print("Edge : Weight\n")
while (no_edge < V - 1):
    # For every vertex in the set S, find the all adjacent vertices
    #(對於集合S中的每個頂點，找到所有相鄰的頂點)
    #, calculate the distance from the vertex selected at step 1.
    #(算出與在步驟1中選擇的頂點)
    # if the vertex is already in the set S, discard it otherwise
    #(如果頂點已經在集合S中，則將其丟棄)
    # choose another vertex nearest to selected vertex  at step 1.
    #(在步驟1選擇另一個與所選頂點最接近的頂點)
    minimum = INF
    x = 0
    y = 0
    for i in range(V):
        if selected[i]:
            for j in range(V):
                if ((not selected[j]) and G[i][j]):  
                    # not in selected and there is an edge
                    if minimum > G[i][j]:
                        minimum = G[i][j]
                        x = i
                        y = j
    print(str(x) + "-" + str(y) + ":" + str(G[x][y]))
    selected[y] = True
    no_edge += 1
```
